```python
import pandas as pd
import tensorflow as tf
import numpy as np

file_path = './W14-01_data.csv'
data = pd.read_csv(file_path)
```

```python
print(data.head())

# 1. Extract the features and labels
X = data[['X1', 'X2']]
Y = data[['Y']]

# 2. Divide train and test
from sklearn.model_selection import train_test_split
X, X_test, Y, Y_test = train_test_split(X, Y, test_size=0.2)

# Print the first 5 rows of the features and labels
print(X.head())
print(Y.head())
```

```python
# Create a Decisopn Tree model
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

tree_clf = DecisionTreeClassifier(random_state=0, criterion='gini', 
                                  splitter='best', max_depth=1 )
tree_clf.fit(X, Y)
```

```python
# Visualize the Decision Tree
import matplotlib.pyplot as plt
plt.figure(figsize=(25,20))
tree.plot_tree(tree_clf, feature_names=['X1', 'X2'], class_names=['0', '1'], filled=True)
plt.show()
```

```python
input = np.array([[0.6717,3.76341], [0.5, 5.5], [0.5, 0.5], [0.5, 1.5]])
output = tree_clf.predict(input)
print(output)

# Test accuracy
from sklearn.metrics import accuracy_score
Y_pred = tree_clf.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
print('Accuracy: ', accuracy)
```
