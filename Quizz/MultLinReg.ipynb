```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
# Ruta al archivo csv
lin_length = 20

data_file = 'quiz01_rdata.csv'

df = pd.read_csv(data_file)

train_size = int(len(df) * 0.7)

#Get 30% of the data for validation and 70% for training
x_train, x_val = df[:train_size], df[train_size:]

# 0 < x1 < 20, 20 < x2 < 40, 40 < x3 < 60
# Filter the DataFrame into three groups
x1 = x_train[(x_train['x'] > 0) & (df['x'] <= lin_length)]
x2 = x_train[(x_train['x'] > lin_length) & (df['x'] <= lin_length * 2)]
x3 = x_train[(x_train['x'] > lin_length * 2) & (df['x'] <= lin_length * 3)]

y1 = x1.pop('y')
y2 = x2.pop('y')
y3 = x3.pop('y')

# Convert the DataFrame and the target column to TensorFlow tensors
y_train_1 = tf.convert_to_tensor(y1.values, dtype=tf.float32, name='y_train_1')
y_train_2 = tf.convert_to_tensor(y2.values, dtype=tf.float32, name='y_train_2')
y_train_3 = tf.convert_to_tensor(y3.values, dtype=tf.float32, name='y_train_3')
x_train_1 = tf.convert_to_tensor(x1.values, dtype=tf.float32, name='x_train_1')
x_train_2 = tf.convert_to_tensor(x2.values, dtype=tf.float32, name='x_train_2')
x_train_3 = tf.convert_to_tensor(x3.values, dtype=tf.float32, name='x_train_3')

y_train_1 = tf.reshape(y_train_1, (-1, 1))
y_train_2 = tf.reshape(y_train_2, (-1, 1))
y_train_3 = tf.reshape(y_train_3, (-1, 1))

x_train_1 = tf.reshape(x_train_1, (-1, 1))
x_train_2 = tf.reshape(x_train_2, (-1, 1))
x_train_3 = tf.reshape(x_train_3, (-1, 1))

x_val1 = x_val[(x_val['x'] > 0) & (df['x'] <= lin_length)]
x_val2 = x_val[(x_val['x'] > lin_length) & (df['x'] <= lin_length * 2)]
x_val3 = x_val[(x_val['x'] > lin_length * 2) & (df['x'] <= lin_length * 3)]

y_val1 = x_val1.pop('y')
y_val2 = x_val2.pop('y')
y_val3 = x_val3.pop('y')

y_val1 = tf.convert_to_tensor(y_val1.values, dtype=tf.float32, name='y_val1')
x_val1 = tf.convert_to_tensor(x_val1.values, dtype=tf.float32, name='x_val1')
y_val2 = tf.convert_to_tensor(y_val2.values, dtype=tf.float32, name='y_val2')
x_val2 = tf.convert_to_tensor(x_val2.values, dtype=tf.float32, name='x_val2')
y_val3 = tf.convert_to_tensor(y_val3.values, dtype=tf.float32, name='y_val3')
x_val3 = tf.convert_to_tensor(x_val3.values, dtype=tf.float32, name='x_val3')

y_val1 = tf.reshape(y_val1, (-1, 1))
y_val2 = tf.reshape(y_val2, (-1, 1))
y_val3 = tf.reshape(y_val3, (-1, 1))

x_val1 = tf.reshape(x_val1, (-1, 1))
x_val2 = tf.reshape(x_val2, (-1, 1))
x_val3 = tf.reshape(x_val3, (-1, 1))

# Print each group
print("Group 1 (0-20):")
print(x_train_1)
print("\nGroup 2 (20-40):")
print(x_train_2)
print("\nGroup 3 (40-60):")
print(x_train_3)
```

```python
# Define the linear regression model
class LinearRegressionModel(tf.Module):
    def __init__(self, input_dim):
        # Initialize weights and bias
        self.weights = tf.Variable(tf.random.normal([input_dim, 1]), dtype=tf.float32)
        self.bias = tf.Variable(tf.zeros([1]), dtype=tf.float32)

    def predict(self, X):
        logits = tf.matmul(X, self.weights) + self.bias
        return logits

    def loss_fn(self, X, y):
        predictions = self.predict(X)
        loss = tf.reduce_mean(tf.square(y - predictions))
        return loss

    def compute_gradients(self, X, y):
        with tf.GradientTape() as tape:
            loss = self.loss_fn(X, y)
        gradients = tape.gradient(loss, [self.weights, self.bias])
        return gradients
```

```python
# Define the training step
def train_step(model, x, y, learning_rate):
    gradients = model.compute_gradients(x, y)
    optimizer = tf.optimizers.SGD(learning_rate)
    optimizer.apply_gradients(zip(gradients, [model.weights, model.bias]))
    return model.loss_fn(x, y)
```

```python
model1 = LinearRegressionModel(input_dim=x_train_1.shape[1])
model2 = LinearRegressionModel(input_dim=x_train_2.shape[1])
model3 = LinearRegressionModel(input_dim=x_train_3.shape[1])

epochs = 1000
lr = 0.0001

model = [model1, model2, model3]
x_train = [x_train_1, x_train_2, x_train_3]
y_train = [y_train_1, y_train_2, y_train_3]

losses_dict = {}

for i in range(x_train.__len__()):
    losses_dict[i] = []
    losses = []
    print(f'Training model {i}')
    for epoch in range(epochs):
        loss = train_step(model[i], x_train[i], y_train[i], learning_rate=lr)
        losses_dict[i].append(loss)
        
        # Monitor gradients
        gradients = model[i].compute_gradients(x_train[i], y_train[i])
        
        
        if epoch % 100 == 0:
            print(f'Epoch {epoch}, Loss: {loss.numpy()}, Learning Rate: {lr}')
            #print(f'Gradients: {gradients}')
```

```python
import matplotlib.pyplot as plt

# Assuming the previous code has been executed and losses_dict is populated

# Plot the losses for each model
for i in range(len(losses_dict)):
    plt.figure(figsize=(10, 6))
    plt.plot(losses_dict[i], label=f'Model {i+1} Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title(f'Loss Curve for Model {i+1}')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import matplotlib.pyplot as plt

# Predict using the trained model (binary output)
y_pred_1 = model[0].predict(x_val1)
y_pred_2 = model[1].predict(x_val2)
y_pred_3 = model[2].predict(x_val3)

y_vals = [y_val1, y_val2, y_val3]
y_preds = [y_pred_1, y_pred_2, y_pred_3]

# Plot the predicted x vs actual y for each model
for i in range(len(y_vals)):
    plt.figure(figsize=(10, 6))
    # Plot predicted x values against actual y values
    plt.scatter(y_preds[i], y_vals[i], label=f'Model {i+1} Predicted X vs Actual Y', alpha=0.5, color='orange')
    plt.plot([min(y_preds[i]), max(y_preds[i])], [min(y_vals[i]), max(y_vals[i])], color='red', linestyle='--')
    plt.xlabel('Predicted X')
    plt.ylabel('Actual Y')
    plt.title(f'Predicted X vs Actual Y for Model {i+1}')
    plt.legend()
    plt.grid(True)
    plt.show()
```
